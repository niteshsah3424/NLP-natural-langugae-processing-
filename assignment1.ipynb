```python
import spacy
import nltk
from nltk.stem import PorterStemmer

nlp = spacy.load("en_core_web_sm")

text = "Apple is looking at buying a startup in Bangalore for $1 billion."
```

```python
# ---------- Tokenization ----------
doc = nlp(text)
tokens = [token.text for token in doc]
print("Tokens:", tokens)
```

```python
# ---------- Stemming (using NLTK) ----------
ps = PorterStemmer()
stems = [ps.stem(token.text) for token in doc]
print("Stems:", stems)
```

```python
# ---------- Lemmatization ----------
lemmas = [(token.text, token.lemma_) for token in doc]
print("Lemmatization:", lemmas)
```

```python
# ---------- POS Tagging ----------
pos_tags = [(token.text, token.pos_, token.tag_) for token in doc]
print("POS Tags:", pos_tags)
```

```python
# ---------- Named Entity Recognition (NER) ----------
ner = [(ent.text, ent.label_) for ent in doc.ents]
print("NER:", ner)
```

```python
import spacy

nlp_hi = spacy.blank("hi")

text_hi = "सोमवार का नाम सुनते ही बैटरी लो हो जाती है।"
doc_hi = nlp_hi(text_hi)

print("Tokens:", [token.text for token in doc_hi])
```

```python
print("Lemmas:", [token.lemma_ for token in doc_hi])
```

```python
print("POS Tags:", [(token.text, token.pos_) for token in doc_hi])
```
