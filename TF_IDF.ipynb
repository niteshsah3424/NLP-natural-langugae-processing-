```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
from sklearn.feature_extraction import text
```

```python
#inatialize tf-idf vectorizer-->create obj
vect=TfidfVectorizer()
```

```python
corpus = [
    "The movie was fantastic and full of suspense",
    "I did not like the movie it was boring",
    "The film had great acting and story",
    "Worst movie I have ever watched"]
corpus
```

```python
x=vect.fit_transform(corpus)
```

```python
x.toarray()
```
